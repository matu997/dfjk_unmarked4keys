メモリ使用量の検証

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
meta_data = pd.read_csv('memory_usage.csv')
data2 = pd.read_csv('memory_usage01.csv')
data3 = pd.read_csv('memory_usage02.csv')
data1 = pd.read_csv('memory_usage00.csv')
data4 = pd.read_csv('memory_usage03.csv')
data5 = pd.read_csv('memory_usage_最大割当.csv')
```

```python

plt.plot(data3['TotalMemory(MB)'], label='desktop Total')
plt.plot(data3['UsedMemory(MB)'], label='desktop Used')

plt.plot(data5['TotalMemory(MB)'], label='IED Total')
plt.plot(data5['UsedMemory(MB)'], label='IED Used')
#x軸にラベルを付ける
plt.xlabel('Time(sec)')
#y軸にラベルを付ける
plt.ylabel('Memory(MB)')
plt.legend()
plt.show()
```

```python
plt.plot(data1['TotalMemory(MB)'], label='1200*1600 Total', color='red')
plt.plot(data1['UsedMemory(MB)'], label='1200*1600 Used', color='green')

#plt.plot(data2['TotalMemory(MB)'])
#plt.plot(data2['UsedMemory(MB)'])

plt.plot(data3['TotalMemory(MB)'], label='300*400 Total', color='blue')
plt.plot(data3['UsedMemory(MB)'], label='300*400 Used', color='orange')

#plt.plot(data4['TotalMemory(MB)'])
#plt.plot(data4['UsedMemory(MB)'])

#x軸にラベルを付ける
plt.xlabel('Time(sec)')
#y軸にラベルを付ける
plt.ylabel('Memory(MB)')
plt.legend()

plt.show()
```
